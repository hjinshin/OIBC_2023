```python
import pandas as pd
import numpy as np
import os
```

```python
gens = pd.read_csv('data/gens.csv')
pred = pd.read_csv('data/pred.csv')
weather_forecast = pd.read_csv('data/weather_forecast.csv')
weather_actual = pd.read_csv('data/weather_actual.csv')
incentive = pd.read_csv('data/incentive.csv')
```

```python
print(pred.shape)
print(gens.shape)
print(weather_forecast.shape)
print(weather_actual.shape)
print(incentive.shape)
```

# 실제 발전량과 예측 발전량

```python
model0_1 = pred[(pred['model_id'] == 0) & (pred['round']==1)]
model0_2 = pred[(pred['model_id'] == 0) & (pred['round']==2)]
model1_1 = pred[(pred['model_id'] == 1) & (pred['round']==1)]
model1_2 = pred[(pred['model_id'] == 1) & (pred['round']==2)]
model2_1 = pred[(pred['model_id'] == 2) & (pred['round']==1)]
model2_2 = pred[(pred['model_id'] == 2) & (pred['round']==2)]
model3_1 = pred[(pred['model_id'] == 3) & (pred['round']==1)]
model3_2 = pred[(pred['model_id'] == 3) & (pred['round']==2)]
model4_1 = pred[(pred['model_id'] == 4) & (pred['round']==1)]
model4_2 = pred[(pred['model_id'] == 4) & (pred['round']==2)]

weather_forecast1 = weather_forecast[weather_forecast['round'] == 1]
weather_forecast2 = weather_forecast[weather_forecast['round'] == 2]

incentive0 =  incentive[incentive['model_id'] == 0]
incentive1 =  incentive[incentive['model_id'] == 1]
incentive2 =  incentive[incentive['model_id'] == 2]
incentive3 =  incentive[incentive['model_id'] == 3]
incentive4 =  incentive[incentive['model_id'] == 4]

model0_1.reset_index(drop=True, inplace=True)
model0_2.reset_index(drop=True, inplace=True)
model1_1.reset_index(drop=True, inplace=True)
model1_2.reset_index(drop=True, inplace=True)
model2_1.reset_index(drop=True, inplace=True)
model2_2.reset_index(drop=True, inplace=True)
model3_1.reset_index(drop=True, inplace=True)
model3_2.reset_index(drop=True, inplace=True)
model4_1.reset_index(drop=True, inplace=True)
model4_2.reset_index(drop=True, inplace=True)
weather_forecast1.reset_index(drop=True, inplace=True)
weather_forecast2.reset_index(drop=True, inplace=True)
incentive0.reset_index(drop=True, inplace=True)
incentive1.reset_index(drop=True, inplace=True)
incentive2.reset_index(drop=True, inplace=True)
incentive3.reset_index(drop=True, inplace=True)
incentive4.reset_index(drop=True, inplace=True)
```

```python
print('gens.shape', gens.shape)
print('model0_1.shape', model0_1.shape)
print('model0_2.shape', model0_2.shape)
print('model1_1.shape', model1_1.shape)
print('model1_2.shape', model1_2.shape)
print('model2_1.shape', model2_1.shape)
print('model2_2.shape', model2_2.shape)
print('model3_1.shape', model3_1.shape)
print('model3_2.shape', model3_2.shape)
print('model4_1.shape', model4_1.shape)
print('model4_2.shape', model4_2.shape)
print('weather_forecast1.shape', weather_forecast1.shape)
print('weather_forecast2.shape', weather_forecast2.shape)
print('incentive0.shape', incentive0.shape)
print('incentive1.shape', incentive1.shape)
print('incentive2.shape', incentive2.shape)
print('incentive3.shape', incentive3.shape)
print('incentive4.shape', incentive4.shape)
```

```python
# model의 round1끼리 비교
diff1 = model0_1[(model0_1['time'] != model1_1['time']) &
                (model0_1['time'] != model2_1['time']) &
                (model0_1['time'] != model3_1['time']) &
                (model0_1['time'] != model4_1['time'])]
# model의 round2끼리 비교
diff2 = model0_2[(model0_2['time'] != model1_2['time']) &
                (model0_2['time'] != model2_2['time']) &
                (model0_2['time'] != model3_2['time']) &
                (model0_2['time'] != model4_2['time'])]
# incentive끼리 비교
diff3 = incentive0[(incentive0['time'] != incentive1['time']) &
                (incentive0['time'] != incentive2['time']) &
                (incentive0['time'] != incentive3['time']) &
                (incentive0['time'] != incentive4['time'])]
# model round2와 incentive 비교
diff4 = model0_2[model0_2['time'] != incentive0['time']]

# model model0_1과 weather_forecast1, weather_actual, gens 비교
diff5 = model0_1[(model0_1['time'] != weather_forecast1['time']) &
                 (model0_1['time'] != weather_actual['time']) &
                 (model0_1['time'] != gens['time'])]
print(diff1.shape, diff2.shape, diff3.shape, diff4.shape, diff5.shape)
```

```python
# weather_forecast2 model_2 병합
merged_df = pd.merge(weather_forecast2, model0_2, on='time', how='inner')
model0_2 = merged_df[['round_y', 'time', 'model_id', 'amount']]
model0_2.rename(columns={
    'round_y': 'round'
}, inplace=True)
model0_2.reset_index(drop=True, inplace=True)
print(model0_2)
```

```python
# model0_2와 다른 model_2들끼리 비교
rows_to_delete = set()
cnt = 0
for i in range(min(model1_2.shape[0], model0_2.shape[0])):
    j = i-cnt
    if model1_2['time'][i] != model0_2['time'][j]:
        rows_to_delete.add(i)
        cnt+=1
print(model0_2.shape, model1_2.shape)
model1_2 = model1_2.drop(index=rows_to_delete)
model2_2 = model2_2.drop(index=rows_to_delete)
model3_2 = model3_2.drop(index=rows_to_delete)
model4_2 = model4_2.drop(index=rows_to_delete)
print(model0_2.shape, model1_2.shape, model2_2.shape, model3_2.shape, model4_2.shape)
print(cnt)
```

```python
# model0_2와 weather_forecast2 비교
rows_to_delete = set()
cnt = 0
for i in range(min(weather_forecast2.shape[0], model0_2.shape[0])):
    j = i-cnt
    if weather_forecast2['time'][i] != model0_2['time'][j]:
        rows_to_delete.add(i)
        cnt+=1
print(model0_2.shape, weather_forecast2.shape)
weather_forecast2 = weather_forecast2.drop(index=rows_to_delete)
print(model0_2.shape, weather_forecast2.shape)
print(cnt)
```

```python
# model0_2와 incentive들끼리 비교
rows_to_delete = set()
cnt = 0
for i in range(min(incentive0.shape[0], model0_2.shape[0])):
    j = i-cnt
    if incentive0['time'][i] != model0_2['time'][j]:
        rows_to_delete.add(i)
        cnt+=1
print(model0_2.shape, incentive0.shape)
incentive0 = incentive0.drop(index=rows_to_delete)
incentive1 = incentive1.drop(index=rows_to_delete)
incentive2 = incentive2.drop(index=rows_to_delete)
incentive3 = incentive3.drop(index=rows_to_delete)
incentive4 = incentive4.drop(index=rows_to_delete)
print(model0_2.shape, incentive0.shape, incentive1.shape, incentive2.shape, incentive3.shape, incentive4.shape)
print(cnt)
```

```python
# model0_2와 model_1들+weather_forecast1들+gens끼리 비교
rows_to_delete = set()
cnt = 0
for i in range(min(model0_1.shape[0], model0_2.shape[0])):
    j = i-cnt
    if model0_1['time'][i] != model0_2['time'][j]:
        rows_to_delete.add(i)
        cnt+=1
print(model0_2.shape, model0_1.shape)
model0_1 = model0_1.drop(index=rows_to_delete)
model1_1 = model1_1.drop(index=rows_to_delete)
model2_1 = model2_1.drop(index=rows_to_delete)
model3_1 = model3_1.drop(index=rows_to_delete)
model4_1 = model4_1.drop(index=rows_to_delete)
weather_forecast1 =weather_forecast1.drop(index=rows_to_delete)
gens =gens.drop(index=rows_to_delete)
print(model0_2.shape, model0_1.shape, model1_1.shape, model2_1.shape, model3_1.shape, model4_1.shape, weather_forecast1.shape, gens.shape)
print(cnt)
```

```python
model0_1.reset_index(drop=True, inplace=True)
model0_2.reset_index(drop=True, inplace=True)
model1_1.reset_index(drop=True, inplace=True)
model1_2.reset_index(drop=True, inplace=True)
model2_1.reset_index(drop=True, inplace=True)
model2_2.reset_index(drop=True, inplace=True)
model3_1.reset_index(drop=True, inplace=True)
model3_2.reset_index(drop=True, inplace=True)
model4_1.reset_index(drop=True, inplace=True)
model4_2.reset_index(drop=True, inplace=True)
weather_forecast1.reset_index(drop=True, inplace=True)
weather_forecast2.reset_index(drop=True, inplace=True)
incentive0.reset_index(drop=True, inplace=True)
incentive1.reset_index(drop=True, inplace=True)
incentive2.reset_index(drop=True, inplace=True)
incentive3.reset_index(drop=True, inplace=True)
incentive4.reset_index(drop=True, inplace=True)
```
